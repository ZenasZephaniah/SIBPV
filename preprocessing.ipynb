```python
# Preprocessing & Feature Engineering
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# Load Data
train_df = pd.read_excel('data/train(1).xlsx')
test_df = pd.read_excel('data/test2.xlsx')
print(f'Train shape: {train_df.shape}')
```

```python
# Feature Engineering: Log Transform Price
# We use log transform to handle the skewness of real estate prices
train_df['price'] = np.log1p(train_df['price'])
```

```python
# Scaling Tabular Data
scaler = StandardScaler()
tab_feats = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'lat', 'long']
train_df[tab_feats] = scaler.fit_transform(train_df[tab_feats])
test_df[tab_feats] = scaler.transform(test_df[tab_feats])
print('Data Scaled successfully.')
```
